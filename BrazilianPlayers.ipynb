```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style = 'darkgrid')
plt.style.use('seaborn-pastel')
```

```python
# Importing and displaying the last 10 players:
df = pd.ExcelFile('br-players.xlsx')
players = pd.read_excel(df, 'br-players')
players.tail(10)
```

```python
# Displaying the number of rows and columns in the dataset:
players.shape
```

```python
# Displaying the datatypes of each column:
players.info()
```

```python
# Displaying players by club:
players['Club'].value_counts().plot(kind = 'barh', color = 'yellow', figsize = (8, 12))
plt.ylabel(None)
plt.title('Players by club', fontweight = 'bold')
plt.show()
```

```python
# Displaying players by country where they play at:
players['Club country'].value_counts().plot(kind = 'barh', color = 'yellow', figsize = (8, 4))
plt.ylabel(None)
plt.title('Players by country where they play at', fontweight = 'bold')
plt.show()
```

```python
# Displaying players by sector:
players['Sector'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Players by sector', fontweight = 'bold')
plt.show()
```

```python
# Displaying players by position:
players['Position'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Players by position', fontweight = 'bold')
plt.show()
```

```python
# Displaying players by region of birth:
players['Region of birth'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Players by region of birth', fontweight = 'bold')
plt.show()
```
