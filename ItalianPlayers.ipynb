```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
from google.colab import drive
from datetime import datetime, date
from dateutil.relativedelta import relativedelta
sns.set(style = 'darkgrid')
plt.style.use('seaborn-pastel')
drive.mount('/content/drive')
```

```python
# Setting the search parameters:
filename = 'it-players.xlsx'
search_path = '/content/drive/My Drive/Datasets/Soccer'

# Searching for the file:
file = None
for root, dirs, files in os.walk(search_path):
    if filename in files:
        file= os.path.join(root, filename)
        break
    df = pd.read_excel(file)
```

```python
# Importing and displaying the last 10 players:
df = pd.ExcelFile(file)
players = pd.read_excel(df, 'it-players')
players.tail(10)
```

```python
# Generating the column 'Age':
now = date.today()
players['Age'] = players.apply(lambda x: relativedelta(now, x['Date of birth']).years if x['Date of birth'] is not pd.NaT else pd.NaT, axis=1)
players.tail(50)
```

```python
# Grouping players by their age group:
def age_groups(row):
  age = row['Age']

  if age < 20:
    return 'Under 20'
  elif age >= 20 and age < 24:
    return 'Between 20 and 23'
  elif age >= 24 and age < 30:
    return 'Between 24 and 30'
  elif age >= 30 and age <= 35:
    return 'Between 30 and 35'
  else:
    return 'Older than 35'

# Applying the function above on a column of the dataset and generating a new column as result:
players['Age group'] = players.apply(age_groups, axis=1)
players.tail()
```

```python
# Displaying players by their age groups:
players['Age group'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Age groups', fontweight = 'bold')
plt.show()
```

```python
# Displaying statistical data of the numerical data present in the dataset:
players.describe()
```

```python
# Splitting the youth clubs by comma and removing leading spaces:
players['Youth club'] = players['Youth club'].str.split(',').apply(lambda x : [i.strip() for i in x])

# Exploding the list into separate rows:
players = players.explode('Youth club')

# Counting Matrix using crosstab method:
club_counts = pd.crosstab(index = players.index, columns = players['Youth club'], margins = False).to_dict('index')
club_counts = pd.DataFrame(club_counts)

# Counting the number of 0s and 1s in each row:
counts = (club_counts.apply(lambda row : [sum(row == 0), sum(row == 1)], axis = 1))
```

```python
# Defining and displaying the final count with 2 columns ('False' and 'True'):
counts = pd.DataFrame(counts.tolist(), index = counts.index, columns = ['False', 'True'])
print(counts)
```

```python
# Displaying the youth clubs that most formed players to play in the NT:
counts['True'].sort_values(ascending = False).plot(kind = 'barh', color = 'green', figsize = (8, 30))
plt.ylabel(None)
plt.title('Youth clubs that most formed players', fontweight = 'bold')
plt.show()
```
