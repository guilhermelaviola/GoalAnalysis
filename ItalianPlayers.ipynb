```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from dateutil.relativedelta import relativedelta
from datetime import datetime, date
sns.set(style = 'darkgrid')
plt.style.use('seaborn-pastel')
```

```python
# Importing and displaying the last 10 players:
df = pd.ExcelFile('br-players.xlsx')
players = pd.read_excel(df, 'br-players')
players.tail(10)
```

```python
# Displaying the number of rows and columns in the dataset:
players.shape
```

```python
# Displaying the datatypes of each column:
players.info()
```

```python
# Displaying players by club:
players['Club'].value_counts().plot(kind = 'barh', color = 'yellow', figsize = (8, 12))
plt.ylabel(None)
plt.title('Players by club', fontweight = 'bold')
plt.show()
```

```python
# Displaying players by country where they play at:
players['Club country'].value_counts().plot(kind = 'barh', color = 'yellow', figsize = (8, 4))
plt.ylabel(None)
plt.title('Players by country where they play at', fontweight = 'bold')
plt.show()
```

```python
# Displaying players by sector:
players['Sector'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Players by sector', fontweight = 'bold')
plt.show()
```

```python
# Displaying players by position:
players['Position'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Players by position', fontweight = 'bold')
plt.show()
```

```python
# Displaying players by region of birth:
players['Region of birth'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Players by region of birth', fontweight = 'bold')
plt.show()
```

```python
# Displaying players by the experience they had in the youth NT:
players['Youth NT experience?'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Youth NT experience?', fontweight = 'bold')
plt.show()
```

```python
# Generating the column 'Age':
now = date.today()
players['Age'] = players.apply(lambda x: relativedelta(now, x['Date of birth']).years if x['Date of birth'] is not pd.NaT else pd.NaT, axis=1)
players.tail(50)
```

```python
# Grouping players by their age group:
def age_groups(row):
  age = row['Age']

  if age < 20:
    return 'Under 20'
  elif age >= 20 and age < 24:
    return 'Between 20 and 23'
  elif age >= 24 and age < 30:
    return 'Between 24 and 30'
  elif age >= 30 and age <= 35:
    return 'Between 30 and 35'
  else:
    return 'Older than 35'

# Applying the function above on a column of the dataset and generating a new column as result:
players['Age group'] = players.apply(age_groups, axis=1)
players.tail()
```

```python
# Displaying players by their age groups:
players['Age group'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Age groups', fontweight = 'bold')
plt.show()
```

```python
# Splitting the youth clubs by comma and removing leading spaces:
players['Youth club'] = players['Youth club'].str.split(',').apply(lambda x : [i.strip() for i in x])

# Exploding the list into separate rows:
players = players.explode('Youth club')

# Counting Matrix using crosstab method:
club_counts = pd.crosstab(index = players.index, columns = players['Youth club'], margins = False).to_dict('index')
club_counts = pd.DataFrame(club_counts)

# Counting the number of 0s and 1s in each row:
counts = (club_counts.apply(lambda row : [sum(row == 0), sum(row == 1)], axis = 1))
```

```python
# Defining and displaying the final count with 2 columns ('False' and 'True'):
counts = pd.DataFrame(counts.tolist(), index = counts.index, columns = ['False', 'True'])
print(counts)
```

```python
# Displaying the youth clubs that most formed players to play in the NT:
counts['True'].sort_values(ascending = False).plot(kind = 'barh', color = 'yellow', figsize = (8, 24))
plt.ylabel(None)
plt.title('Youth clubs that most formed players', fontweight = 'bold')
plt.show()
```
