```python
# Importing all the necessary libraries:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style = 'darkgrid')
plt.style.use('seaborn-pastel')
```

```python
# Importing and displaying the last 10 goals scored:
df = pd.ExcelFile('Non-Bayern Bundesliga Champions.xlsx')
champions = pd.read_excel(df, 'champions')
champions.tail(10)
```

I extracted this data from Transfermarkt.

```python
# Displaying the number of rows and columns in the dataset:
champions.shape
```

```python
# Displaying the datatypes of each column:
champions.info()
```

```python
# Displaying champions by shirt number:
champions['Number'].value_counts().plot(kind = 'barh', color = 'red', figsize = (8, 16))
plt.ylabel(None)
plt.title('Champions by Number', fontweight = 'bold')
plt.show()
```

```python
# Displaying champions by position:
champions['Position'].value_counts().plot(kind = 'barh', color = 'red', figsize = (8, 4))
plt.ylabel(None)
plt.title('Champions by Position', fontweight = 'bold')
plt.show()
```

```python
# Displaying champions by nation:
champions['Nation'].value_counts().plot(kind = 'barh', color = 'red', figsize = (8, 16))
plt.ylabel(None)
plt.title('Champions by Nation', fontweight = 'bold')
plt.show()
```

```python
# Displaying champions by club:
champions['Club'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Champions by Club', fontweight = 'bold')
plt.show()
```

```python
# Displaying champions by year:
champions['Year'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Champions by Year', fontweight = 'bold')
plt.show()
```

```python
# Displaying leadership proportion (captains, vice-captains and none):
champions['Leadership'].value_counts().plot(kind = 'pie', autopct = '%.0f%%', figsize = (8, 8), wedgeprops = dict(width = 0.4), pctdistance = 0.8)
plt.ylabel(None)
plt.title('Leadership', fontweight = 'bold')
plt.show()
```
